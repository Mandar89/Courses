```python
# Data Cleaning on entire dataset 
# Let's define a pipeline to clean up all the messages 
# The pipeline performs the following: (1) remove punctuation, (2) remove stopwords

def message_cleaning(message):
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

```python
# Let's test the newly added function
reviews_df_clean = reviews_df['verified_reviews'].apply(lambda x: message_cleaning(str(x)))
```

```python
print(reviews_df_clean[3]) # show the cleaned up version
```

```python
print(reviews_df['verified_reviews'][3]) # show the original version
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Define the cleaning pipeline we defined earlier
reviews_df['verified_reviews'] = reviews_df['verified_reviews'].fillna('')
vectorizer = CountVectorizer(analyzer = message_cleaning)
reviews_countvectorizer = vectorizer.fit_transform(reviews_df['verified_reviews'])
```

```python
print(vectorizer.get_feature_names_out())
```

```python
print(reviews_countvectorizer.toarray()) 
```

```python
reviews_countvectorizer.shape
```

```python
reviews_df
```

```python
# first let's drop the column
reviews_df.drop(['verified_reviews'], axis=1, inplace=True)
reviews = pd.DataFrame(reviews_countvectorizer.toarray())
```

```python
# Now let's concatenate them together
reviews_df = pd.concat([reviews_df, reviews], axis=1)
```

```python
reviews_df
```

```python
# Let's drop the target label coloumns
X = reviews_df.drop(['feedback'],axis=1)
```

```python
y = reviews_df['feedback']
```

```python
X.shape
```

```python
y.shape
```

```python
# Splitting the dataset into Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Applying the Naive Bayes model to the Training set
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
X_train.columns = X_train.columns.astype(str)
NB_classifier.fit(X_train, y_train)
```

```python
# Checking the Confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Predicting the Training set results
y_predict_train = NB_classifier.predict(X_train)
```

```python
y_predict_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
```

```python
sns.heatmap(cm, annot=True)
```

```python
# Predicting the Test set results
X_test.columns = X_test.columns.astype(str)
y_predict_test = NB_classifier.predict(X_test)
```

```python
y_predict_test
```

```python
cm = confusion_matrix(y_test, y_predict_test)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```

```python
# Applying Logistic Regression Classifier on the Training set
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Testing Set Performance
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print('Accuracy {} %'.format( 100 * accuracy_score(y_pred, y_test)))
```
